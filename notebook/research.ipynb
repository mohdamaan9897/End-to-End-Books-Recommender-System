```python
import pandas  as pd
import numpy as np
```

```python
books = pd.read_csv('BX-Books.csv', sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
books.head()
```

```python
books.iloc[10]['Image-URL-L']
```

```python
books.shape
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.head()
```

```python
books.rename(columns={'Book-Title': 'title', 'Book-Author': 'author', 'Year-Of-Publication': 'year',"Publisher":'publisher',"Image-URL-L":"image_url"},inplace=True)
```

```python
books.head()
```

```python
users = pd.read_csv('BX-Users.csv', sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
users.head()
```

```python
users.shape
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
```

```python
# Now Load the third dataframe

ratings = pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
```

```python
print(books.shape, users.shape, ratings.shape,sep='\n')
```

```python
ratings['user_id'].value_counts()
```

```python
# Lets store users who have rated more than 200 books
x = ratings['user_id'].value_counts()>200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
# Now join ratings with books
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```
